```python
import pandas as pd
import numpy as np
```

```python
poi=pd.read_csv("landmarks-and-places-of-interest-including-schools-theatres-health-services-spor.csv")
poi
```

```python
poi.loc[poi['Sub Theme'].str.contains('Retail', case=False, na=False), 'Sub Theme'] = 'Retail'
```

```python
poi.loc[poi['Theme'] == 'Mixed Use', 'Theme'] = 'Retail'
poi
```

```python
poi.loc[poi['Sub Theme'].str.contains('Department Store', case=False, na=False), 'Sub Theme'] = 'Retail'
```

```python
poi["Theme"].unique()
```

```python
filtered_poi = poi[(poi['Theme'] == "Leisure/Recreation") | (poi['Theme'] == "Place of Worship")|(poi['Theme'] == "Place Of Assembly")|(poi['Theme'] == "Retail")|(poi['Theme'] == "Community Use")|(poi['Theme'] == "Purpose Built")]
```

```python
filtered_poi['lat'], filtered_poi['lon'] = filtered_poi['Co-ordinates'].str.split(',', 1).str
```

```python
filtered_poi=filtered_poi[["Theme","Sub Theme","Feature Name","lat","lon"]]
filtered_poi.to_csv("filtered_poi.csv",index=False)
filtered_poi
```

```python
filtered_poi.loc[filtered_poi['Theme'].str.contains('Purpose Built', case=False, na=False), 'Theme'] = 'Place Of Assembly'
filtered_poi.to_csv("filtered_poi.csv",index=False)
filtered_poi
```
